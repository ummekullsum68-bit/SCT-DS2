```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')

df = pd.read_csv(r'C:\Users\yashw\OneDrive\Desktop\Projects\SCT DS 2\Data\train.csv')

print("First 5 Rows:")
print(df.head())

print("\nDataset Shape:")
print(df.shape)

print("\nDataset Info:")
print(df.info())

print("\nStatistical Summary:")
print(df.describe())

print("\nMissing Values:")
print(df.isnull().sum())

df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df.drop(columns='Cabin', inplace=True)

print("\nMissing Values After Cleaning:")
print(df.isnull().sum())


df.drop_duplicates(inplace=True)

print("\nDataset Shape After Removing Duplicates:")
print(df.shape)


plt.figure(figsize=(6,4))

sns.countplot(x='Survived', data=df)

plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')

plt.savefig(r'C:\Users\yashw\OneDrive\Desktop\Projects\SCT DS 2\Image/survival_count.png')

plt.show()


plt.figure(figsize=(6,4))

sns.countplot(x='Sex', data=df)

plt.title('Gender Distribution')

plt.savefig(r'C:\Users\yashw\OneDrive\Desktop\Projects\SCT DS 2\Image\gender_distribution.png')

plt.show()


plt.figure(figsize=(8,5))

sns.histplot(df['Age'], bins=20, kde=True)

plt.title('Age Distribution')

plt.savefig(r'C:\Users\yashw\OneDrive\Desktop\Projects\SCT DS 2\Image\age_distribution.png')

plt.show()


plt.figure(figsize=(6,4))

sns.countplot(x='Pclass', data=df)

plt.title('Passenger Class Distribution')

plt.savefig(r'C:\Users\yashw\OneDrive\Desktop\Projects\SCT DS 2\Image\passenger_class.png')

plt.show()


numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(10,6))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')

plt.savefig(r'C:\Users\yashw\OneDrive\Desktop\Projects\SCT DS 2\Image\correlation_heatmap.png')

plt.show()


print("\n========== Insights ==========")

print("1. Missing values handled successfully.")
print("2. Cabin column dropped due to excessive missing values.")
print("3. Most passengers were male.")
print("4. Majority of passengers belonged to class 3.")
print("5. Survival distribution visualized successfully.")

print("\nTask 2 Completed Successfully!")
```
